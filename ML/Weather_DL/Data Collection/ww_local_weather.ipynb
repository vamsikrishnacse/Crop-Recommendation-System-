```python
import datetime
m_d=[31,28,31,30,31,30,31,31,30,31,30,31]
s_d=1
s_m=1
s_y=2019
e_d=m_d[s_m-1]
e_m=s_m
e_y=2019

end_date=""
year=[]
while(end_date !='01-Jun-2020'):
    x_s = datetime.datetime(s_y, s_m, s_d)
    x_start=x_s.strftime("%d-%b-%Y")
    print(e_y, e_m, e_d)
    x_e=datetime.datetime(e_y, e_m, e_d)
    x_end=x_e.strftime("%d-%b-%Y")
    print("Start"+x_start+" End : "+x_end)
    URL="https://api.worldweatheronline.com/premium/v1/past-weather.ashx"

    CITY="Tiruchirappalli"
    KEY="5b5bf18d10f248999a3162113200911"
    PARAMS={"key":KEY,"q":CITY,"date":x_start,"enddate":x_end,"tp":24,"format":"json"}
    result= requests.get(url = URL, params = PARAMS) 
    y = result.json()
    data=y["data"]
    weather=data["weather"]
    year.append(weather)
    s_m+=1
    if(s_m==13):
        s_m=1
        s_y+=1
        e_y=s_y
    e_m=s_m
    e_d=m_d[e_m-1]
    end_date=x_start

```

```python
cleaned_data=[]
for i in range(len(year)):
    for j in range(len(year[i])):
        desc=year[i][j]["hourly"][0]['weatherDesc'][0]["value"]
        date=year[i][j]["date"]
        temp_f=year[i][j]["hourly"][0]["tempF"]
        temp_c=year[i][j]["hourly"][0]["tempC"]
        hum=year[i][j]['hourly'][0]['humidity']
        pres=year[i][j]['hourly'][0]['pressure']
        cleaned_data.append([date,desc,temp_f,temp_c,hum,pres])
```

```python
import pandas as pd

df=pd.DataFrame(cleaned_data,columns=["Date","Weather","TempF","TempC","Humidity","Pressure"],index=None)
```
