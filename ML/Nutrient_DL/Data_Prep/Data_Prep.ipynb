```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("nutrients.csv")
```

```python
details=[]
for i in range(data.shape[0]):
    crop,n,p,k,soil=data.iloc[i]
    print(crop,n,p,k,soil)
    n_range=(max(n-3,0),n+1.5)
    p_range=(max(p-1,0),p+1)
    k_range=(max(k-0.5,0),k+1)
    details.append([crop,n_range,p_range,k_range,soil])
details
```

```python
rows=[]
count=1
for c in range(len(details)):
    for i in range(200):
        n=np.round(np.random.uniform(low=details[c][1][0],high=details[c][1][1]),decimals=3)
        p=np.round(np.random.uniform(low=details[c][2][0],high=details[c][2][1]),decimals=3)
        k=np.round(np.random.uniform(low=details[c][3][0],high=details[c][3][1]),decimals=3)
        rows.append([count,details[c][0],n,p,k,details[c][4]])
        count+=1
```

```python
rows[0:202]
```

```python
df=pd.DataFrame(rows,columns=["index","Crop_Name","N","P","K","soil_type"])
```

```python
df.to_csv("Nutrients_table.csv")
```
